#*2024년 국가별 음식물 쓰레기 발생량 내림차순 그래프*
(내림차순)

```python
# prompt: /content/global_food_wastage_dataset.csv 에서 country 랑 Total Waste (Tons) 에 해당하는 값을 가지고, 막대그래프로 나타내줘.Year는 2024년으로 하고, 정렬은 Total Waste (Tons) 값이 큰 순서부터 해줘. 그래프의 제목은 "Countries with most food wast in 2024"로 해줘.

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/global_food_wastage_dataset.csv')

# 'Country'와 'Total Waste (Tons)' 열만 선택
df = df[['Country', 'Total Waste (Tons)']]

# Year 열을 추가하고 2024로 설정
df['Year'] = 2024

# Total Waste (Tons) 값을 기준으로 내림차순 정렬
df = df.sort_values('Total Waste (Tons)', ascending=False)


# 막대 그래프 그리기
plt.figure(figsize=(15, 8))
plt.bar(df['Country'], df['Total Waste (Tons)'])
plt.xlabel('Country')
plt.ylabel('Total Waste (Tons)')
plt.title('Countries with most food wast in 2024')
plt.xticks(rotation=90)  # x축 레이블 회전
plt.tight_layout()
plt.show()

```

#*우리나라의 음식 항목 별 음식물 쓰레기 발생량*
(내림차순)

```python
# prompt: /content/global_food_wastage_dataset.csv 에서 South Korea 행에서 food category 열의 내용을 막대그래프로 만들어줘. 내림차순으로 정리해 줘

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/global_food_wastage_dataset.csv')

# South Korea 행 선택
south_korea_data = df[df['Country'] == 'South Korea']

# food category 열 선택
food_categories = south_korea_data['Food Category']

# 각 food category의 빈도수 계산
food_category_counts = food_categories.value_counts()

# 내림차순으로 정렬
food_category_counts = food_category_counts.sort_values(ascending=False)

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(food_category_counts.index, food_category_counts.values)
plt.xlabel('Food Category')
plt.ylabel('Count')
plt.title('Food Category Counts in South Korea')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

우리나라에서 발생하는 음식물 쓰레기중 가장 많은 양을 차지하는 것은 *냉동식품* 인 것을 알 수 있다.

#우리나라의 연도별 음식물 쓰레기로 인한 금액 손실

```python
# prompt: /content/global_food_wastage_dataset.csv 에서 south korea 행의 데이터를 사용하여, 가로변은 Year, 세로변은 Economic Loss (Million $)로 하여금 막대프래프를 만들어줘. 정렬은 연도별 오름차순으로

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/global_food_wastage_dataset.csv')

# South Korea 행 선택
south_korea_data = df[df['Country'] == 'South Korea']

# Year와 Economic Loss (Million $) 열만 선택
south_korea_data = south_korea_data[['Year', 'Economic Loss (Million $)']]

# Year를 기준으로 오름차순 정렬
south_korea_data = south_korea_data.sort_values('Year')

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(south_korea_data['Year'], south_korea_data['Economic Loss (Million $)'])
plt.xlabel('Year')
plt.ylabel('Economic Loss (Million $)')
plt.title('Economic Loss from Food Wastage in South Korea')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

우리나라의 음식물 쓰레기 발생량은 대체적으로 감소하고 있는 추세인 것을 알 수 있다.

```python
# prompt: /content/global_food_wastage_dataset.csv에서 한국과 미국의 연도별 발생하는 냉동식품 음식물 쓰레기의 양을 비교하는 그래프를 그려줘.

import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('/content/global_food_wastage_dataset.csv')

# 한국 데이터만 추출
korea_data = df[(df['Country'] == 'South Korea') & (df['Food Category'] == 'Frozen Food')]

# 미국 데이터만 추출
usa_data = df[(df['Country'] == 'USA') & (df['Food Category'] == 'Frozen Food')]

# 연도별 데이터 그룹화 및 총량 계산
korea_waste_by_year = korea_data.groupby('Year')['Total Waste (Tons)'].sum()
usa_waste_by_year = usa_data.groupby('Year')['Total Waste (Tons)'].sum()

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(korea_waste_by_year.index, korea_waste_by_year.values, label='South Korea')
plt.plot(usa_waste_by_year.index, usa_waste_by_year.values, label='USA')
plt.xlabel('Year')
plt.ylabel('Total Waste (Tons)')
plt.title('Comparison of Frozen Food Waste between South Korea and USA')
plt.legend()
plt.grid(True)
plt.show()
```

2019년과 2023년 사이에서 우리나라에서 발생하는 냄동식품 음식물쓰레기의 양이
미국에서 발생하는 냉동식품 음식물 쓰레기의 양을 상회하였다는 것을 알 수 있다.

2019년은 *COVID-19* 감염병이 발생하였던 때로,
직전 연도인 2018년도와 비교해서 우리나라의 냉동식품 소비량이 이 때문에 증가한 것이라 예상 할 수 있다.
